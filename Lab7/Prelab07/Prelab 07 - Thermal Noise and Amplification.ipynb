Let $X$ be a Gaussian random variable with mean $\mu$ and variance $\sigma^2$. Evaluate the probabilities of $\mu - \sigma \leq X \leq \mu + \sigma$ and of $\mu - 3\sigma \leq X \leq \mu + 3\sigma$, using the $Q$ function.

*ANSWER:*

1)
Normalize random variable $X' = \frac{X-\mu}{\sigma}$

Normalize range to get new bounds $R1 = \frac{(\mu-\sigma)-\mu}{\sigma} = 1$

Because it is symmetric, the range is -1 < X < 1

Q function give the tail of the distribution for its input. To chop of each end, we subtract it twice from the total probability, 1:

$P(-1 < X' < 1) = 1 - 2Q(1)$

Similarly for the case of bounds $\mu - 3\sigma \leq X \leq \mu + 3\sigma$, the normalization works out:


$P(-3 < X' < 3) = 1 - 2Q(3)$

The first two components of a typical RF front end are a band-pass filter and a low-noise amplifier. The band-pass filter is a passive circuit and hence, a lossy component (gain less than one). The order of cascade of noiseless ideal components is not important. This problem explores the consequences of ordering when the internal noise produced by each component is taken into account.

#### Question 7.4a:

Consider the cascade of the band-pass filter followed by the low-noise amplifier. Let the gains of the two be -6 dB and 30 dB, respectively. Also, let the noise figures of the two be 3 dB and 0.6 dB, respectively. Determine the noise figure of this ordering.

*ANSWER:*

$NF_{12} = NF_1 + \frac{NF_2 - 1}{G_1}$

NF_1 = 3 dB = 1.995

NF_2 = 0.6 dB = 1.148

G_1 = -6 dB = 0.251

G_2 = 30 dB = 1000

$NF_{12} = 1.995 + \frac{1.148 - 1}{0.251} = 2.585 = 4.12$ dB

#### Question 7.4b:

Now consider the cascade of the same two components, but in reverse order. Determine the noise figure of this ordering.

*ANSWER:*

$NF_{21} = NF_2 + \frac{NF_1 - 1}{G_2}$

NF_1 = 3 dB = 1.995

NF_2 = 0.6 dB = 1.148

G_1 = -6 dB = 0.251

G_2 = 30 dB = 1000

$NF_{21} = 1.148 + \frac{1.995 - 1}{1000} = 1.149 = 0.6$ dB

### Exercise 7.6: Sample Mean and Sample Variance of a Gaussian Distribution

In this exercise, we will explore behavior of parameter estimators as a function of the number of measurements obtained.

#### Question 7.6a:

Write a Python code (in the code cells below) to calcuate the sample mean and sample variance (biased) [Wiki: Biased Sample Variance](https://en.wikipedia.org/wiki/Variance) of number of samples of a Gaussian random variable with mean $\mu = 0$ and variance $\sigma^2 = 100$. Your code by run seven times in which the number of generated samples vary from $10^{\{3, 4, \cdots, 9\}}$, respectively, and should calculate and print the sample mean and sample variance for each of the seven trials.

*Note:* Use data type numpy.float32 instead of the default numpy.float64 to reduce the time required for execution of the code.

*ANSWER:*

[...]

```python
import numpy as np

num_samp = np.array([1e3, 1e4, 1e5, 1e6, 1e7, 1e8, 1e9], dtype = np.int32) # number of samples for each trial
mean = 0
var = 100
samp_mean = np.zeros(len(num_samp), dtype = np.float32)
samp_var = np.zeros(len(num_samp), dtype = np.float32)

for i in range(len(num_samp)):
    samples = np.random.default_rng().normal(mean, np.sqrt(var), num_samp[i]) # generate 10^{i + 3} samples of a Gaussian random variable using numpy functions (use datatype np.float32)
    samp_mean[i] = np.mean(samples)# calculate the sample mean
    samp_var[i] = np.var(samples)# calculate the biased sample variance
    

## add code to print out your values
print("Sample Means:")
for mean_i in samp_mean: print(mean_i)
print("\nSample Variances:")
for var_i in samp_var: print(var_i)
```

#### Question 7.6b:

What do you observe about the behavior of the sample mean and sample variance, as the number of samples increases?


*ANSWER:*

As the number of samples increases, the calculated sample mean and sample variance converges closer to the expected values.
