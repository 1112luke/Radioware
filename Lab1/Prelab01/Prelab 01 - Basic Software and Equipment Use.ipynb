### Exercise 1.2: Generating and Plotting a Continuous-Time Sinusoid

In this exercise, we will generate samples of a cosine signal and plot them. We will also develop a "rule of thumb" for how much oversampling is helpful for the standard plotting algorithm, which uses linear interpolation.

Suppose we want to plot a continuous-time (CT) cosine signal
$$x(t) = A \cos(2\pi f_0 t),$$
where $f_0$ is the (fundamental) frequency. We can achieve this by computing discrete-time (DT) samples of the signal with sampling frequency $f_s$, or sampling period $T_s = 1/f_s$, as
$$x[n] = x(nT_s)=A\cos(2\pi f_0 n T_s)=A\cos(2\pi u_0 n),$$
where the normalized discrete time frequency is $u_0 = f_0 / f_s$.

The lines of code below define specific values for the amplitude $A$, frequency $f_0$, sampling frequency $f_s$ as an integer multiple $M$ of the Nyquist frequency $2f_0$, and compute the samples $x[n]$ for an integer $N$ periods of the sinusoid.

````python
A = 2.0
f_0 = 100.0
T_0 = 1/f_0

M = 3
f_s = M*2.0*f_0
T_s = 1/f_s

N = 5
t_max = N*T_0

t=np.linspace(0,t_max,int(t_max/T_s),endpoint=False)
x=A*np.cos(2*np.pi*f_0*t)
````

The lines of code below plot this signal in Matplotlib using linear interpolation, i.e., connecting successive DT samples of the CT signal with straight lines.

````python
plt.plot(t,x,'b.',t,x,'b-')
plt.legend(['Samples','Linear Interpolation'])
plt.xlabel('Time $t$ (seconds)')
plt.ylabel('Signal $x(t)$')
````

When the ADALM2000 samples a signal, which we will denote by $x_1(t)$, it first feeds it into an anti-alias filter $H_1(f)$ to remove possible out-of-band signals. Then, it converts the filtered continuous signal into a discrete one. This can be represented as

$x_{3}[n] = x_{2}(nT_{\text{ADC}})$,

where $T_{\text{ADC}} = \frac{1}{f_{\text{ADC}}}$ is the sample period of the ADC. 

In many applications, the received signal is downsampled (or decimated) to reduce the complexity of the software processing. Downsampling or decimating a discrete signal by a factor of $M_1$ consists of keeping every $M_1$th value of that signal. In our case, $x_{3}[n]$ is downsampled to $x_{4}[n]$:

$x_{4}[n] = x_{3}[nM_1]$.

After some algebraic manipulations and following the lecture notes, we can obtain the expression below for discrete-time downsampling:
$$X_{4}(e^{j2\pi u_r})=\frac{1}{M_1}\sum_{m-0}^{M_1-1}X_{3}(e^{j2\pi (u_r-m)/M_1}),$$
where $u_r = \frac{f}{f_{\text{ADC}}}$. This relationship is illustrated in the figure below. We observe that the spectra of the sampled and decimated sequences differ in frequency scaling.

### Exercise 1.8: DAC and Interpolation

Generating waveforms from samples starts with a Python script that sends a sequence of samples to the ADALM2000 via a USB cable. The ADALM2000 generates waveforms on its output ports at a fixed rate $f_{\text{DAC}}$. To reduce processing complexity, the effective sampling rate of the samples sent to the ADALM2000 is typically much lower than $f_{\text{DAC}}$. That is why we need an interpolation stage: to upsample the signal to reach $f_{\text{DAC}}$.

To better grasp the advantage of interpolation, let's assume that the signal of interest is a 1 kHz sinusoid. It could be generated and then sent to the ADALM with a sampling rate of $f_{\text{DAC}}$, but that would be wasteful since $f_{\text{DAC}}$ is typically on the order of mega (one million) samples per second. Instead, we produce the waveform at a reasonable frequency (e.g., 3000 samples per second) and then the ADALM interpolates it to make it suitable for transmission at $f_{\text{DAC}}$.

### Exercise 1.9: DAC Reconstruction Filter and Equivalent Representation

We now want to look more closely at the DAC reconstruction filter and its equivalent representation.

The AD9963 chip has a DAC that converts the discrete (or digital) signal $x_7[n]$ into a continuous-time signal $x_{9}(t)$. We can model this in two steps, first, by converting $x_7[n]$ into a continuous-time impulse train.

$x_8(t) = \sum\limits_{k=-\infty}^{k=+\infty} x_7[k] \delta(t-k)$,

and then by convolving $x_{8}(t)$ with a rectangular impulse response, $h_2(t)$, commonly refered to as a zero-order hold function.

$x_9(t) = x_8(t) * h_2(t)$.

As shown in the figure below, the zero-hold function is not an ideal band-limited interpolation filter and may pass some energy from the image: $X_{9}(f)$ contains images that lie beyond $f_{\text{DAC}}$. Therefore, another analog lowpass filter is used to suppress these images.
